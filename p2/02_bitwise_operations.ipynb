```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### **Cargar las imÃ¡genes y pre-procesamiento**
---

```python
def binarize_img(image, threshold):
  newImg = np.where(image > threshold, 255, 0)
        
  return newImg
```

```python
def resize_img(image, new_width, new_height):
  resized_image = cv2.resize(image, (new_width, new_height))
        
  return resized_image
```

```python
path_1 = './img1.jpg'
height = 300
width = 400


img_1_color = cv2.cvtColor(cv2.imread(path_1), cv2.COLOR_BGR2RGB)
resz_img_1_color = resize_img(img_1_color, width, height)


# Display the image using matplotlib
plt.imshow(resz_img_1_color)
```

```python
# Uso opencv para cargar la imagen y pasarla a escala de grises

img_1 = cv2.imread(path_1, cv2.IMREAD_GRAYSCALE)
resz_img_1 = resize_img(img_1, width, height)
plt.imshow(resz_img_1, cmap='gray')
```

```python
binarized_1 = binarize_img(resz_img_1, 40)
plt.imshow(binarized_1, cmap='gray')
```

```python
path_2 = './img2.jpg'
img_2_color = cv2.cvtColor(cv2.imread(path_2), cv2.COLOR_BGR2RGB)
resz_img_2_color = resize_img(img_2_color, width, height)


# Display the image using matplotlib
plt.imshow(resz_img_2_color)
```

```python
img_2 = cv2.imread(path_2, cv2.IMREAD_GRAYSCALE)
resz_img_2 = resize_img(img_2, width, height)
plt.imshow(resz_img_2, cmap='gray')
```

```python
binarized_2 = binarize_img(resz_img_2, 60)
plt.imshow(binarized_2, cmap='gray')
```

### **Operaciones aritmÃ©ticas y lÃ³gicas**
---

```python
def image_addition(colored_img1, colored_img2, height, width):
  newImg = np.zeros((height, width, 3), dtype=np.uint8)
  
  for i in range(height):
    for j in range(width):
       for c in range(3):  # Para cada canal (R, G, B)
          sum_value = int(colored_img1[i, j, c]) + int(colored_img2[i, j, c])
          newImg[i, j, c] = min(sum_value, 255)
  return newImg
```

```python
addition = image_addition(resz_img_1_color, resz_img_2_color, height, width)
plt.imshow(addition)
```

```python
def image_subtraction(colored_img1, colored_img2, height, width):
  newImg = np.zeros((height, width, 3), dtype=np.uint8)
  
  for i in range(height):
    for j in range(width):
       for c in range(3):  # Para cada canal (R, G, B)
          sub_value = int(colored_img1[i, j, c]) - int(colored_img2[i, j, c])
          newImg[i, j, c] = max(sub_value, 0)
  return newImg
```

```python
subtraction = image_subtraction(resz_img_1_color, resz_img_2_color, height, width)
plt.imshow(subtraction)
```

```python
def image_bitwise_and(img1, img2, height, width):
  newImg = np.zeros((height, width, 1))
  
  for i in range(height):
    for j in range(width):
      newImg[i, j] = img1[i, j] & img2[i, j]
  
  return newImg
```

```python
or_bitwise = image_bitwise_and(binarized_1, binarized_2, height, width)
plt.imshow(or_bitwise, cmap='gray')
```

```python
def image_bitwise_or(img1, img2, height, width):
  newImg = np.zeros((height, width, 1))
  
  for i in range(height):
    for j in range(width):
      newImg[i, j] = img1[i, j] | img2[i, j]
  
  return newImg
```

```python
or_bitwise = image_bitwise_or(binarized_1, binarized_2, height, width)
plt.imshow(or_bitwise, cmap='gray')
```

```python
def image_bitwise_xor(img1, img2, height, width):
  newImg = np.zeros((height, width, 1))
  
  for i in range(height):
    for j in range(width):
      newImg[i, j] = img1[i, j] ^ img2[i, j]
  
  return newImg
```

```python
xor_bitwise = image_bitwise_xor(binarized_1, binarized_2, height, width)
plt.imshow(xor_bitwise, cmap='gray')
```
