```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### **Cargar la imagen**
---

```python
# Uso opencv para cargar la imagen y pasarla a escala de grises
path = './labradors.jpg'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

### **Solución "a mano"**
---

```python
def binarize_img(image, threshold):
  # np.where applies some function or value to each element of a vector / matrix 
  # based on a condition (Offers better performance than 2 nested loops)
  
  # for i in range(height):
  #   for j in range(width):
      
  newImg = np.where(image > threshold, 255, 0)
        
  return newImg
      
```

```python
binarized = binarize_img(img, 90)
plt.imshow(binarized, cmap='gray')
```

### **Solución usando funciones de opencv**
---

```python
def binarize_img_cv2(image, threshold):
  _, global_threshold = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY, )
  
  plt.imshow(global_threshold, cmap='gray')
      
```

```python
binarize_img_cv2(img, 90)
```
