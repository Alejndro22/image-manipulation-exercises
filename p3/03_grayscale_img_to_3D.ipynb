```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import cv2
```

```python
# Uso opencv para cargar la imagen y pasarla a escala de grises
path = './CR7.jpg'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
def plot_grayscale_img_3D(img, elev = False, azim = False):
  x = np.arange(img.shape[1])
  y = np.arange(img.shape[0])
  x, y = np.meshgrid(x, y)
  z = img

  
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  ax.plot_surface(x, y, z, cmap='gray')
  
  if elev and azim:
    ax.view_init(elev, azim)
      
  return 
```

```python
plot_grayscale_img_3D(img)
```

```python
# Una vista un poco más aerea
plot_grayscale_img_3D(img, 270, 60)
```

```python
# Una vista un totalmente aerea
plot_grayscale_img_3D(img, 270, 90)
```
